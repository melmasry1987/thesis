```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
from sklearn import tree

x = train.drop(['H1DiskTime'], axis = 1)
y = train['H1DiskTime']

model = tree.DecisionTreeRegressor(max_depth=5)

model.fit(x,y)
y.head()
```

```python
import graphviz

feature_names = list(train.drop(['H1DiskTime'], axis=1))

dot_data = tree.export_graphviz(model, out_file=None, filled=True, rounded=True,
                                feature_names=feature_names)
graph = graphviz.Source(dot_data)  
graph
```

```python
from sklearn.tree import _tree

def rule(tree, feature_names):
    tree_ = tree.tree_
    feature_name = [
        feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!"
        for i in tree_.feature
    ]
    print("def tree({}):".format(", ".join(feature_names)), file = open('Network_Rules.txt','w'))

    def recurse(node, depth):
        indent = "  " * depth
        if tree_.feature[node] != _tree.TREE_UNDEFINED:
            name = feature_name[node]
            threshold = tree_.threshold[node]
            print("{}if {} <= {}:".format(indent, name, threshold), file = open('Network_Rules.txt','a'))
            recurse(tree_.children_left[node], depth + 1)
            print("{}else:  # if {} > {}".format(indent, name, threshold), file = open('Network_Rules.txt','a'))
            recurse(tree_.children_right[node], depth + 1)
        else:
            print("{}return {}".format(indent, tree_.value[node]), file = open('Network_Rules.txt','a'))

    recurse(0, 1)

rule(model, list(train.drop(['H1DiskTime'], axis=1)))
```

```python
predict = model.predict(test)
```

```python
from pandas import DataFrame

output = DataFrame({'H1DiskTime': predict})
output.to_csv('Network_Output.csv', index = False)
```
